```python
# ========================== VENTILATION DES LIGNES DU DATAFRAME PAR DATES FIXES COMPRISES ENTRE '2013-01-01 00:00:00' et  '2022-12-31 23:30:00' PAR PAS DE 00H30

from tqdm import tqdm

# Conversion des colonnes 'date_debut' et 'date_fin_previsionnelle' en datetime
ind_prod_comp['date_debut'] = pd.to_datetime(ind_prod_comp['date_debut'])
ind_prod_comp['date_fin_previsionnelle'] = pd.to_datetime(ind_prod_comp['date_fin_previsionnelle'])

# Création de la liste de dates incrémentées de 30 minutes
liste_dates = pd.date_range(start='2013-01-01 00:00:00', end='2022-12-31 23:30:00', freq='30min')

# Création de la liste pour stocker les nouvelles lignes
new_rows = []

# Parcours de chaque ligne de ind_prod_comp
for idx, row in tqdm(ind_prod_comp.iterrows(), total=ind_prod_comp.shape[0]):
    date_debut = row['date_debut']
    date_fin_previsionnelle = row['date_fin_previsionnelle']

    # Détermination de la borne inférieure et supérieure des dates
    borne_inf_index = liste_dates.searchsorted(date_debut, side='left')
    borne_sup_index = liste_dates.searchsorted(date_fin_previsionnelle, side='right')

    # Vérification que les indices sont dans les limites
    if borne_inf_index >= len(liste_dates):
        borne_inf_index = len(liste_dates) - 1
    if borne_sup_index >= len(liste_dates):
        borne_sup_index = len(liste_dates) - 1

    # Création d'une série de dates comprises entre borne_inf et borne_sup
    dates_range = liste_dates[borne_inf_index:borne_sup_index]

    # Copie de la ligne avec toutes les dates dans new_rows
    new_row = row.drop(['date_debut', 'date_fin_previsionnelle']).to_dict()
    for date in dates_range:
        new_row['date'] = date
        new_rows.append(new_row.copy())

# Création  d'un DataFrame à partir de la liste de nouvelles lignes
df_result_ind_prod = pd.DataFrame(new_rows)

# Sauvegarde du DataFrame résultant dans un fichier CSV
output_path = 'df_result_ind_prod.csv'
df_result_ind_prod.to_csv(output_path, index=False)

output_path


print("FIN CODE VENTILATION")
```
